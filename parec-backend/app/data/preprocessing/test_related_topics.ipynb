## Testing our trained Top2Vec model

- what terms are most similar to a query?
- which similar topics are suggested?
- which wordclouds can be generated for visualization?


```python
from top2vec import Top2Vec
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load the saved model
model = Top2Vec.load('t2v_lemmatized_with_stats')
```

```python
# Define an input string
input_str = 'artificial intelligence'

print("Numbers of topics found by Top2Vec: ", model.get_num_topics())

topics_words, _, _ = model.get_topics()
print("Topics ordered by size: ", topics_words)
```

```python
model.generate_topic_wordcloud(10, background_color='black', reduced=False)
```

```python
related_words, _ = model.similar_words(keywords=["deep neural"], num_words=10)
```

```python
#for words learned by the model:
topic_words, word_scores, topic_scores, topic_nums = model.search_topics(keywords=["machine learning"], num_topics=5)
```

```python
first_topic = topic_words[0][:5]
second_topic = topic_words[1][:5]
third_topic = topic_words[2][:5]
forth_topic = topic_words[3][:5]
fith_topic = topic_words[4][:5]

print(first_topic, second_topic, third_topic, forth_topic, fith_topic)
```
