```python
from top2vec import Top2Vec
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load the saved model
model = Top2Vec.load('.././data/t2v_with_n_grams_and_args')
```

```python
# Define the input string
input_str = 'machine learning'

print("Numbers of topics found by Top2Vec: ", model.get_num_topics())

topics_words, _, _ = model.get_topics()
print("Topics ordered by size: ", topics_words)
```

```python
model.generate_topic_wordcloud(3, background_color='black', reduced=False)
```

```python
print("Top words per topic: ", model.topic_words[1])
```

```python
topics_based_on_query, _, _, _ = model.query_topics(input_str, num_topics=1, reduced=False, tokenizer=None)
print("Words closest to input: ", topics_based_on_query)
```

```python
topic_words, word_scores, topic_scores, topic_nums = model.query_topics(input_str, num_topics=5)
```

```python
# generate the topic wordcloud
topic_words, word_scores, topic_scores, topic_nums = model.search_topics(keywords=["reinforcement learning"], num_topics=1)
for topic in topic_nums:
    model.generate_topic_wordcloud(topic)
```
